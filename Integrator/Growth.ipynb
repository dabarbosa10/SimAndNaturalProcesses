```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
s0=0
t0=1
deltaT=0.05
```

```
def impl(s0,t0,tf,deltaT):
  N=int((tf-t0)/deltaT)
  t=np.arange(t0,tf,deltaT)
  vec=np.zeros(N)
  vec[0]=s0
  for i in range(1,N):
    vec[i]=vec[i-1]-10*deltaT*(vec[i-1])
  return vec,t
```

```
v1,t1=impl(1.0,0,10,0.05)
print(v1)
```

```
plt.plot(t1,v1,'.')
```

```
def growth(t0,tf,v,C, P0,DeltaT):

  N=int((tf-t0)/DeltaT)
  t=np.arange(t0,tf,DeltaT)
  vec=np.zeros(N)
  vec[0]=P0

  for i in range(1,N):
    vec[i]=vec[i-1]+DeltaT*(v*(1-vec[i-1]/C)*vec[i-1])

  return vec,t
```

```
v2,t2=growth(0,9,1,50,10,0.01)
```

```
vec3,tim3=growth(0,9,1,50,90,0.01)
```

```
plt.figure(figsize=(10,8))
plt.plot(t2,v2,'b.')
plt.plot(tim3,vec3,'r.')
plt.hlines(50,0,9,colors='k',linestyles='dashed')
plt.xlabel('$t$',fontsize=24)
plt.ylabel('$P(t)$',fontsize=24)
plt.xlim(0,9)
plt.ylim(10,90)
plt.tick_params(labelsize=20, pad=6)
plt.grid()
plt.show()
```
