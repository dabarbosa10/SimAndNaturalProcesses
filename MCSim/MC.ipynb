```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
def coin_flip():
    return random.randint(0,1)
```

```python
coin_flip()
```

```python
list1=[]
def monte_carlo(n):
    results =0
    for i in range(n):
        flip_result= coin_flip()
        results=results+flip_result
        prob_value=results/(i+1)
        list1.append(prob_value)
        
        plt.axhline(y=0.5,color='r',linestyle='-')
        plt.xlabel("Iterations")
        plt.ylabel("Probability")
        plt.plot(list1)
    return results/n
```

```python
result=monte_carlo(5000)
```

```python
def monte_carlo1(n):
    pro=[]
    head=0
    tail=0
    for i in range(n):
        fresult=coin_flip()
        if fresult==0:
            head+=1
            pro.append(head/(i+1))
        else:
            pro.append(pro[i-1])
    return np.array(pro)
```

```python
answer=monte_carlo1(5000)
```

```python
plt.figure(figsize=(13,7))
plt.axhline(y=0.5,color='r',linestyle='-')
plt.xlabel("Iterations")
plt.ylabel("Probability")
plt.grid()
plt.plot(answer,'bd-')
```
