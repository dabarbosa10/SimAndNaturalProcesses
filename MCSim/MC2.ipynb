```python
import matplotlib.pyplot as plt
import numpy as np
import random
%matplotlib inline
```

```python
doors=["goat","goat","car"]
```

```python
def Monty_Hall(n):
    switch_winsP=[]
    stick_winsP=[]
    switch_P=0
    stick_P=0
    for i in range(n):
        random.shuffle(doors)
        resp=random.randrange(2)
        if doors[resp] !='car':
            switch_P+=1
        else:
            stick_P+=1
        switch_winsP.append(switch_P/(i+1))
        stick_winsP.append(stick_P/(i+1))
    return switch_winsP, stick_winsP
```

```python
change,keep=Monty_Hall(4000)
```

```python
plt.figure(figsize=(13,7))
plt.axhline(y=0.6666,color='r',linestyle='-')
plt.axhline(y=0.3333,color='g',linestyle='-')
plt.xlabel('Iterations')
plt.ylabel('Probability')
plt.plot(change,'b',label='Change Door wins')
plt.plot(keep,'k',label='Sitck Response Looses')
plt.legend()
plt.grid()
```
