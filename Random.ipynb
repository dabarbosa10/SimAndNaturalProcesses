```python
#Let's think of a particle collection that can move either to the left or the right. This 
#movement is driven by a flip coin.
import random 
import numpy as np
import matplotlib.pyplot as plt
num_par=4
num_steps=2000
pos=np.zeros(num_par)
HEAD=1; TAIL=2
```

```python
for i in range(num_steps):
    plt.plot(i,pos[0],'r*')
    plt.plot(i,pos[1],'kd')
    plt.plot(i,pos[2],'b*')
    plt.plot(i,pos[3],'gd')
    for p in range(num_par):
        coin=random.randint(1,2)
        if coin==HEAD:
            pos[p]+=1
        elif coin==TAIL:
            pos[p]-=1
plt.grid()    
```
